```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
df = pd.read_csv("netflix_data.csv")
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df["type"].head(20))
```

```python
print(df["type"].value_counts())
```

```python
# Selecting only "Movie" in "type" DataFrame

movie = df[df["type"] == "Movie"]
print(len(movie))
```

```python
# What was the most frequent movie duration in the 1990s? Save an approximate answer as an integer called duration (use 1990 as the decade's start year).

movie1990 = movie[(movie["release_year"] >= 1990) & (movie["release_year"] < 2000)]
movie1990.value_counts()
```

```python
# Visualization of duration column in the filtered data.

plt.hist(movie1990["duration"], edgecolor = "black")
plt.xlabel("Duration")
plt.ylabel("Number of Movies")
plt.title("Duration of Movies in the 1990s")
plt.show()
```

A movie is considered short if it is less than 90 minutes. Count the number of short action movies released in the 1990s and save this integer as short_movie_count.

```python
# Filter the "action" movies in the genre
action_movies= movie1990[movie1990["genre"] == "Action"]
action_movies.head()
```

```python
short_movie_count = action_movies[action_movies["duration"] < 90].value_counts()
print(len(short_movie_count))
```

```python
# Another way

short_movie_count = 0
for label, row in action_movies.iterrows():
    if row["duration"] < 90:
        short_movie_count = short_movie_count + 1
    else:
      short_movie_count = short_movie_count
print(short_movie_count)
```
